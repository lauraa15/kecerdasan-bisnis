1. Pre rpocessing dan nomer 1

```python
# importings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# load data
df = pd.read_csv('GermanCredit.csv')
```

```python
#cek isi data
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Cek distribusi Target (Good vs Bad)
plt.figure(figsize=(6, 4))
sns.countplot(x='RESPONSE', data=df, hue='RESPONSE', legend=False, palette='viridis')
plt.title('Distribusi Kelas Target (0: Bad, 1: Good)')
plt.xlabel('Credit Response')
plt.ylabel('Jumlah Nasabah')
plt.show()
```

Dari chart diatas, data tidak terlalu seimbang. Sekitar 70% adalah good credit (1) dan 30% bad credit (0). Karena ini categorical variable maka perlu diubah menjadi angka.

2. Nomor 2 / split data dan train model

```python
# membagi fitur (x) dan target (y)
X = df.drop('RESPONSE', axis=1)
y = df['RESPONSE']
print("Data Features:", X.shape)
```

```python
# categorical data diubah jadi angka (One-Hot Encoding)
X = pd.get_dummies(X, drop_first=True)
```

```python
# split data (60% train, 40% val)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.4, random_state=42)
print(f"Jumlah Data Training: {X_train.shape[0]}")
print(f"Jumlah Data Validation: {X_val.shape[0]}")
```

```python
# Train model logistic regression
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
# train model decision tree
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
```

```python
# train model neural network
nn_model = MLPClassifier(max_iter=1000, random_state=42)
nn_model.fit(X_train, y_train) 
```

3. Nomor 3 / Hitung profit atau evaluasi model

```python
def calculate_net_profit(model, X_val, y_val, model_name):
    y_pred = model.predict(X_val)
    acc=accuracy_score(y_val, y_pred)
    cm = confusion_matrix(y_val, y_pred)
    
    tn, fp, fn, tp = cm.ravel()
    
    # hitung Profit
    profit = (tp * 100) - (fp * 500)
    
    print(f"--- {model_name} ---")
    print(f"Accuracy: {acc*100:.2f}%")
    print(f"Confusion Matrix:\n{cm}")
    print(f"True Positive (Untung 100): {tp}")
    print(f"False Positive (Rugi 500) : {fp}")
    print(f"Total Net Profit          : {profit} DM")
    print("-" * 30)
    
    return profit
```

```python
profit_log  = calculate_net_profit(log_model, X_val, y_val, "Logistic Regression")
profit_tree = calculate_net_profit(tree_model, X_val, y_val, "Decision Tree")
profit_nn   = calculate_net_profit(nn_model, X_val, y_val, "Neural Network")
```

Hasil perhitungan net profit dan accuracy model adalah:
1. Logistic regression memiliki akurasi paling tinggi (77.25%), dengan Net Profit -7,300 DM
2. Decision tree memiliki akurasi 69.5% dengan profit -10,800 DM
3. Neural Network memiliki akurasi 61.75% dengan profit -9,900

Highest profit dihasilkan oleh model Logistic regression dengan profit -7,300.

Dari sini kita bisa melihat bahwa ketiga model meskipun memiliki akurasi yang lumayan bagus, namun hasil dari net profit sangatlah minus (merugi) dan model belum sesuai harapan. 

4. Nomor 4 / penyesuaian model

```python
# memilih nasabah terbaik
# menggunakan logistic regression

# ambil probabilitas nasahab baik (Good)
y_prob = log_model.predict_proba(X_val)[:, 1]

# Merge data asli untuk sorting
result_df = pd.DataFrame({
    'Actual': y_val, 
    'Prob_Good': y_prob
})
```

```python
# sort descending dari probability good tertinggi
result_df = result_df.sort_values(by='Prob_Good', ascending=False).reset_index(drop=True)
```

```python
# itung profit kumulatif
# actual 1 (Good) -> Profit +100
# actual 0 (Bad) -> Profit -500
result_df['Profit'] = result_df['Actual'].apply(lambda x: 100 if x == 1 else -500)
result_df['Cumulative_Profit'] = result_df['Profit'].cumsum()
```

```python
# Plot Chart (Profit Curve)
plt.figure(figsize=(10, 6))
plt.plot(result_df['Cumulative_Profit'])
plt.xlabel('Jumlah Nasabah yang Disetujui (Sorted by Score)')
plt.ylabel('Total Net Profit (DM)')
plt.title('Profit Lift Chart')
plt.grid(True)
plt.show()
```

```python
# Cutoff terbaik
max_profit = result_df['Cumulative_Profit'].max()
best_cutoff_index = result_df['Cumulative_Profit'].idxmax()
best_probability_threshold = result_df.loc[best_cutoff_index, 'Prob_Good']

print(f"Profit max: {max_profit} DM")
print(f"Index max/cutoff: {best_cutoff_index}")
print(f"Cutoff probaility: {best_probability_threshold:.4f}")
```

Maka, profit tertinggi mencapai profit 8,400 DM, yakni di nasabah ke 173 sebelum akhirnya grafik profit meluncur ke bawah. Untuk probability cutoff ada di 0.8233 atau 82.33% yang berarti nasabah dengan credit skor di bawah 82.33% harus ditolak agar tidak merugi
