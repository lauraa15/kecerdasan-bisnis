```python
import pandas as pd
import scipy.stats
import numpy as np

control_df = pd.read_csv("control_group.csv", sep=";")
test_df = pd.read_csv("test_group.csv", sep=";")
```

```python
print(control_df.head())
print(test_df.head())
```

```python
medianspend_control = np.median(control_df['Spend [USD]'])
medianspend_test = np.median(test_df['Spend [USD]'])
```

```python
print("Medianspend Control Group:", medianspend_control)
print("Medianspend Test Group:", medianspend_test)

print(scipy.stats.ttest_ind(control_df['Spend [USD]'], test_df['Spend [USD]']))
```

```python
print(np.mean(test_df['Spend [USD]'])-np.mean(control_df['Spend [USD]']))
```

```python
# cohen d
cohen_d = (np.mean(test_df['Spend [USD]']) - np.mean(control_df['Spend [USD]'])) / np.std(control_df['Spend [USD]'])
print("Cohen's d:", cohen_d)

if abs(cohen_d) < 0.2:
    print("Effect size: Small")
elif abs(cohen_d) < 0.5:
    print("Effect size: Medium")
else:
    print("Effect size: Large")
```

```python
from statsmodels.stats.power import TTestIndPower
alpha = 0.05
nobs = min(len(control_df), len(test_df))
           
power_analysis = TTestIndPower()
power = power_analysis.solve_power(effect_size=cohen_d,nobs1 = len(control_df),alpha=alpha)
print("Power: ",power)
```
